```python
from selenium import webdriver
import time
from lxml import etree
```

```python
# 实例化一个chrome浏览器对象
browser = webdriver.Chrome()

# 打开url地址（此处的get不是requests中的get请求）
browser.get('https://www.epubit.com/books')

print(browser.page_source) #看看是否能获取完整的网页源码
```

```python


# 能获取网页源码后，我们可以爬取想要的数据了
def extract_data(page_source):
    tree = etree.HTML(page_source)
    book_list = tree.xpath('//*[@id="bookItem"]/a')
    for i in book_list:
        book_name = i.xpath('./div[@class="list-title moreline2"]/text()')[0]
        book_price = i.xpath('./div[@class="list-price clear"]/div[@class="price fl"]/text()')[0]
        print(book_name+'\n'+book_price)
        print('=========================================================================')

with open('page_source.html','w',encoding='utf-8')as f: #保存到本地看看是否是完整的html
    f.write(browser.page_source)
    
# 定位下一页标签
def to_next_page():
    next_page = browser.find_element_by_xpath('//*[@id="entry"]/div[2]/div[2]/div[2]/div/button[2]')
    next_page.click()

def get_data():
    # 获取10页数据
    for i in range(10):
        extract_data(browser.page_source)
        time.sleep(1)
        to_next_page()
        time.sleep(2)


get_data() # 运行爬虫
browser.quit() # 关闭浏览器
```
