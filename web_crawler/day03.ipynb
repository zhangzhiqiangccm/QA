```python
import requests
from lxml import etree
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.105 Safari/537.36'
}
```

```python
# 将爬取的过程封装到函数
def get_movies_info(url):
    response = requests.get(url=url, headers=headers)
    tree = etree.HTML(response.text)
    rank_num = tree.xpath('//div[@class="item"]/div[1]/em/text()')
    titles = tree.xpath('//div[@class="item"]/div[2]/div[1]/a/span[1]/text()')
    score = tree.xpath('//div[@class="item"]/div[2]/div[2]/div/span[2]/text()')
    movie_info = list(zip(rank_num, titles, score))
    return movie_info
```

```python
movie_list = []
# 保存爬取每一页的电影信息
def run_douban250_spider():
    # 使用for循环执行十次
    n = 0
    for i in range(1,11):
        print(f'开始爬取，第{i}页！')
        # 翻页从start=0开始，每次翻页后n+25...直到start=250
        url = f'https://movie.douban.com/top250?start={n}&filter='
        # 调用上面封装好的函数并将每页电影信息保存到movie_list列表中
        movie_list.append(get_movies_info(url))
        n += 25
    print('爬取完成！！展示结果：')
    print(movie_list)
```

```python
# 调用爬虫函数
run_douban250_spider()
```
