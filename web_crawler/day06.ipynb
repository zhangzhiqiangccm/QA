```python
from lxml import etree
import requests

header = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.141 Safari/537.36',
}

# 随便挑一部小说的目录页url
url = 'https://www.tsxs.org/16/16814/'

def get_name(url):
    r = requests.get(url)
    r.encoding = 'gbk'
    tree = etree.HTML(r.text)
    title = tree.xpath('// *[@id="maininfo"]/div[1]/h1/text()')[0]
    print(f'小说的名字是{title}')
    return title
```

```python
def mkfolder(folder_name):
    if not os.path.exists(folder_name):
        print('文件夹不存在，先创建文件夹')
        os.makedirs(folder_name)
    # 切换到文件件中
    os.chdir(folder_name)

name = get_name(url)
print(name)
mkfolder(f'./{name}')
```

```python
def get_catalogue(url):
    r = requests.get(url)
    r.encoding = 'gbk'
    tree = etree.HTML(r.text)
    title = tree.xpath('//*[@id="chapterlist"]/li/a/text()')
    link = tree.xpath('//*[@id="chapterlist"]/li/a/@href')
    catalog = list(zip(title, link))
    print(catalog)
    return catalog

base_url = 'https://www.tsxs.org/'
name = get_name(url)
print(name)
mkfolder(f'./{name}')
print(get_catalogue(url))
```
